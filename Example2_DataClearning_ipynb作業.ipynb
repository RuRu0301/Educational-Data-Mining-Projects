```python
import json
# 把json欄位轉換成dataframe
df_expanded = pd.json_normalize(df['json'].apply(json.loads))
df_expanded
```

```python
import re
import numpy as np

# 把整份資料中的 NaN 補 0
df_expanded = df_expanded.fillna(0)

# 時間轉成總秒數
def time_str_to_seconds(text):
    if isinstance(text, (int, float)):  # 如果已經是數字就直接回傳
        return text
    if not isinstance(text, str):
        return 0.0

    hours = minutes = seconds = 0

    hour_match = re.search(r'(\d+)\s*時', text)
    minute_match = re.search(r'(\d+)\s*分', text)
    second_match = re.search(r'(\d+)\s*秒', text)

    if hour_match:
        hours = int(hour_match.group(1))
    if minute_match:
        minutes = int(minute_match.group(1))
    if second_match:
        seconds = int(second_match.group(1))

    return float(hours * 3600 + minutes * 60 + seconds)

# 尋找欄位名稱中包含「時間」兩字的欄位
time_columns = [col for col in df_expanded.columns if '時間' in col]

# 對每個欄位套用轉換函數
for col in time_columns:
    df_expanded[col] = df_expanded[col].apply(time_str_to_seconds)

df_expanded
```

```python
# 找出重複的學號
duplicated_students = df_expanded[df_expanded.duplicated('stuid', keep=False)]
duplicated_students
```

```python
# 找出 Null 的位置
null_mask = df_expanded['stuid'].isnull()
missing_count = df_expanded['stuid'].isnull().sum()
print(f"缺失值筆數：{missing_count}")
null_string_mask = df_expanded['stuid'] == 'Null'
num_null_string = null_string_mask.sum()
print(f"缺失字串: {num_null_string}") #Null其實是字串
```

```python
# 取得現有的 ID 集合，避免新 ID 重複
existing_ids = set(df_expanded['stuid'].unique())

# 產生新 ID（如 missing_001, missing_002, ...）
new_ids = []
counter = 1
while len(new_ids) < num_null_string:
    new_id = f"missing_{counter:03d}"
    if new_id not in existing_ids:
        new_ids.append(new_id)
    counter += 1

# 替換掉 "Null" 為新 ID（依照順序填回去）
df_expanded.loc[null_string_mask, 'stuid'] = new_ids

# 確保全部是字串（一致性）
df_expanded['stuid'] = df_expanded['stuid'].astype(str)

len(set(df_expanded['stuid']))==len(df_expanded['stuid']) #沒有重複的學號
```

```python
print(df_expanded.columns.tolist())
df_selected = df_expanded[['stuid','本月素材總使用時間','本月瀏覽影片時間','本月測驗時間','本月語音作業時間','本月音檔時間','本月書籍瀏覽時間','本月課間教材時間']]
df_selected
```

根據每位學生的瀏覽日期，來去畫出不同學生瀏覽影片的天數

```python
# 排除 stuid 欄位，對其他欄加總
sums = df_selected.drop(columns='stuid').sum()

# 畫圖
import matplotlib.pyplot as plt
import plotly.express as px


# 將 Series 轉換成 DataFrame（因為 Plotly 需要表格格式）
sum_df = sums.reset_index()
sum_df.columns = ['活動類型', '總秒數']

# 若想轉換成「小時」單位
sum_df['總時數'] = sum_df['總秒數'] / 3600

# 使用 Plotly 繪圖（x 軸為活動類型，y 軸為總時數）
fig = px.bar(sum_df,
             x='活動類型',
             y='總時數',
             labels={'活動類型': '活動類別', '總時數': '總時數（小時）'},
             title='各類學習活動的總時數分佈')

# 顯示圖表
fig.show()
```

```python
# 排除「本月素材總使用時間」欄位
sums_others = sums.drop(labels='本月素材總使用時間')

# 計算比例
percentages = sums_others / sums_others.sum() * 100
percent_df = percentages.reset_index()
percent_df.columns = ['活動類型', '百分比']

# 圓餅圖
fig = px.pie(percent_df,
             names='活動類型',
             values='百分比',
             title='各學習活動時間占比（不含本月素材總使用時間）')

fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 建立子圖：1 列 2 欄（左右並排），指定第二個是 pie chart
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=("活動百分比 - 長條圖", "活動百分比 - 圓餅圖"),
                    specs=[[{"type": "bar"}, {"type": "pie"}]])

# 加入長條圖（百分比）
fig.add_trace(
    go.Bar(x=percent_df['活動類型'], y=percent_df['百分比'], name="百分比"),
    row=1, col=1
)

# 加入圓餅圖
fig.add_trace(
    go.Pie(labels=percent_df['活動類型'], values=percent_df['百分比'], name="百分比"),
    row=1, col=2
)

# 圖表標題與版面設定
fig.update_layout(
    title_text="各學習活動時間百分比分布（不含本月素材總使用時間）",
    height=500, width=1000,
    showlegend=False
)

fig.show()
```

在上述的數據清理與分析過程中，利用了來自宜蘭教網中心數據資料，2023/12學習吧使用數據，並針對整體學生該月使用學習吧的使用時間及比例進行探討。

首先，透過 Google Sheets API，我們將資料載入 Pandas DataFrame 進行清理:先把所有空值補0後，再將時間相關欄位資料型態轉換成數值型資料，並轉換成秒方便接下來的計算與視覺化。

接著提取欲分析的欄位stuid,本月素材總使用時間,本月瀏覽影片時間,本月測驗時間,本月語音作業時間,本月音檔時間,本月書籍瀏覽時間,本月課間教材時間，並通過 Plotly 繪製的長條圖及圓餅圖。

根據資料，發現學生最主要使用的功能是語音作業功能，這是出乎我意料的，因此推測教師使用學習吧派出了許多語音作業。此外，瀏覽影片、測驗、書籍瀏覽時間也占了相當高比例，又以影片觀看時間略高於閱讀時間，推測學生比起閱讀更喜歡用影片觀看來學習。而音檔和課間教材則無使用或幾乎無使用。

此結果可以幫助教育管理者了解學生學習型態，檢討並進一步優化平台內容與功能，以提升學生的學習體驗和成效。
